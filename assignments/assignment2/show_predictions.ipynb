# Visualize Predictions

Labels predicted with unet with input 512 pixels, trained on 10000 batches.

```python
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
def get_paths(img, example_dir='./example'):
    base = os.path.splitext(img)[0]
    
    img = os.path.join(example_dir, 'images', base + '.jpg')
    label = os.path.join(example_dir, 'labels', base + '.png')
    prediction = os.path.join(example_dir, 'predictions', base + '.png')

    return img, label, prediction
```

```python
paths = [get_paths(img) for img in os.listdir('./example/images')]
palette_image = Image.open('./example/labels/00-QP-J4lW-IMLqGyHb5pQ.png')
```

```python
for img, label, prediction in paths:
    fig = plt.figure(figsize=(18, 18))

    # Original image
    img = Image.open(img)
    plt.subplot(1, 3, 1)
    plt.imshow(img)
    
    # Label
    label = Image.open(label)
    plt.subplot(1, 3, 2)
    plt.imshow(label)
    
    # Prediction
    prediction = Image.open(prediction).quantize(palette=palette_image)
    plt.subplot(1, 3, 3)
    plt.imshow(prediction)
```
