```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data

from model import CNNModel
from visualizer import FilterVisualizer
```

```python
model = CNNModel('./models/model1')
data = input_data.read_data_sets('MNIST_data/')
```

```python
sess = tf.Session()
viz = FilterVisualizer(sess, model, data)
```

```python
filter_imgs = viz.visualizeLayer(model.conv1, 3)
```

```python
fig=plt.figure(figsize=(16, 16))
columns = len(list(filter_imgs[0]))
rows = len(filter_imgs)
for i, imgs in enumerate(filter_imgs):
    for j, img in enumerate(imgs):
        fig.add_subplot(rows, columns, i * columns + j+ 1)
        plt.imshow(img)
        
```
