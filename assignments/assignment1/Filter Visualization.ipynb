```python
import matplotlib.pyplot as plt
import tensorflow as tf

from model import CNNModel
from visualizer import FilterVisualizer

tf.logging.set_verbosity(tf.logging.ERROR)

from tensorflow.examples.tutorials.mnist import input_data
```

```python
model = CNNModel('model1')
data = input_data.read_data_sets('MNIST_data/')
```

```python
sess = tf.Session()
model.load(sess)
viz = FilterVisualizer(sess, model, data)
```

```python
def plot_filter_imgs(filter_imgs):
    rows = len(filter_imgs)
    columns = len(filter_imgs[0])
    fig = plt.figure(figsize=(16, 16))
    
    for i, imgs in enumerate(filter_imgs):
        for j, img in enumerate(imgs):
            fig.add_subplot(rows, columns, i * columns + j + 1)
            plt.imshow(img)
```

## Network Architecture
![](./media/img1.png)

### 1st Convolutional Layer

```python
# window (conv(3x3))-> (3x3)
plot_filter_imgs(viz.visualizeLayer(model.conv1, 3))
```

### 2nd Convolutional Layer

```python
# window (conv(3x3)=>conv(3x3))-> (5x5)
plot_filter_imgs(viz.visualizeLayer(model.conv2, 5))
```

### 3rd Convolutional Layer

```python
# window (conv(3x3)=>max_pool(2x2, stride=2)=>conv(3x2)=>conv(3x3))-> (13x13)
plot_filter_imgs(viz.visualizeLayer(model.conv3, 13))
```

### 4th Convolutional Layer

```python
# window (conv(3x3)=>conv(3x3)=>max_pool(2x2, stride=2)=>conv(3x2)=>conv(3x3))-> (13x13)
plot_filter_imgs(viz.visualizeLayer(model.conv4, 29))
```
